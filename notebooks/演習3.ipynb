```python
import numpy as np
import networkx as nx
from graph_theory.utils import plot_graph, incicence2adjacency, exists_graph
```

## 演習3.1

```python
mat_1 = np.zeros((5, 5))
g_1 = nx.from_numpy_matrix(mat_1, parallel_edges=True, create_using=nx.MultiGraph())
plot_graph(g_1)
```

```python
mat_2 = np.full((6, 6), 1)
mat_2[range(6), range(6)] = 0
g_2 = nx.from_numpy_matrix(np.triu(mat_2), parallel_edges=True, create_using=nx.MultiGraph())
plot_graph(g_2, seed=3)
```

```python
arr1 = np.repeat(np.array([[0, 0, 1, 1, 1, 1]]), 2, axis=0)
arr2 = np.repeat(np.array([[1, 1, 0, 0, 0, 0]]), 4, axis=0)
mat_3 = np.r_[arr1, arr2]

g_3 = nx.from_numpy_matrix(np.triu(mat_3), parallel_edges=True, create_using=nx.MultiGraph())
plot_graph(g_3, seed=3)
```
